```python
def partition(x, low, high): 
    i = (low-1)     
    pivot = x[high]   
    for j in range(low, high): 
        if x[j] <= pivot: 
            i = i+1
            x[i], x[j] = x[j], x[i]
    x[i+1], x[high] = x[high], x[i+1] 
    return (i+1) 
  
def quickSort(x, low, high): 
    if len(x) == 1: 
        return x 
    if low < high: 
        pi = partition(x, low, high) 
        quickSort(x, low, pi-1) 
        quickSort(x, pi+1, high) 
    return x

def selectie(x):
    n=len(x)
    for i in range(n-1):
        k=i
        for j in range(i+1,n):
            if x[k]>x[j]:
                k=j
        if k!=i:
            x[i]+=x[k]
            x[k]=x[i]-x[k]
            x[i]-=x[k]
    return x
    
def insertie(x):
    n=len(x)
    for i in range(1,n):
        aux=x[i]
        j=i-1
        while j>=0 and aux<x[j]:
            x[j+1]=x[j]
            j-=1
        x[j+1]=aux
    return x

def bubbleSort(x):
    n=len(x)
    t=n-1
    while t>0:
        m=t
        t=-1
        for i in range(m):
            if x[i]>x[i+1]:
                x[i]+=x[i+1]
                x[i+1]=x[i]-x[i+1]
                x[i]-=x[i+1]
                t=i
    return x

def mergeSort(x):
    n=len(x)
    if n>1:
        m=n//2
        L=x[:m]
        R=x[m:]
        mergeSort(L)
        mergeSort(R)
        lL=len(L) 
        lR=len(R)
        i=0
        j=0
        k=0
        while i < lL and j < lR:
            if L[i] < R[j]:
                x[k] = L[i]
                i += 1
            else:
                x[k] = R[j]
                j += 1
            k += 1

        while i < lL:
            x[k] = L[i]
            i += 1
            k += 1

        while j < lR:
            x[k] = R[j]
            j += 1
            k += 1
    return x

def countingSort(x):
    m=max(x)
    c=[0 for i in range(m+1)]
    for i in x:
        c[i]+=1
    x=[]
    for i in range(m+1):
        while c[i]>0:
            x.append(i)
            c[i]-=1
    return x

```
